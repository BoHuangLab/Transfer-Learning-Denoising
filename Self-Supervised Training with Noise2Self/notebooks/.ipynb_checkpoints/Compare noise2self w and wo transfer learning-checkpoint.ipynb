# Run results
performance vs sample_size

```python
config = {
    sample_size_list : [1] + [i for i in range(5,51,5)],
    root : '/home/yina/DeepDenoising/Data/Confocal/',
    psignal_levels : [30], #[s for s in range(10,100,10)],
    types : ['microtubule'], # ['DNA', 'lysosome', 'microtubule', 'mitochondria'],
    captures : 1,
    train_stat : True,
    patch_size : 128,
    batch_size : 32,
    n_iters : 300,
    repeats : 10,
    metrics : ['mse', 'ssmi', 'frc']
}
```

```python
sample_size_list = [1] + [i for i in range(5,51,5)]
```

```python
root = '/home/yina/DeepDenoising/Data/Confocal/'
loader = tiff_loader
psignal_levels = [30] #[s for s in range(10,100,10)]
types = ['microtubule'] # ['DNA', 'lysosome', 'microtubule', 'mitochondria']
captures = 1
train_stat=True
patch_size = 128
batch_size = 32
```

```python
FMD = {}
random = {}
```

```python
from metric import quantify
from train import train
```

```python
n_iters = 300
repeats = 10
metrics_key = ['mse', 'ssmi', 'frc']

for sample_size in sample_size_list:
    key_s = sample_size
    FMD[key_s] = {}
    random[key_s] = {}
    print(f"Running training with {sample_size} sample_size...")
    n_epoch = int(n_iters / math.ceil(sample_size / batch_size))
    
    for repeat in range(repeats):
        print(f"No. {repeat}...")
        loader, dataset = load_confocal(root, train_stat, batch_size, psignal_levels, sample_size, split_ratio = 0.2, types=types, captures=captures,
                      patch_size=patch_size, loader=tiff_loader)
        noisy = dataset[0]
        clean = dataset[1]
        noisy, clean = noisy.type(torch.FloatTensor).to(device), clean.type(torch.FloatTensor).to(device)

        model_random = Unet()
        model_FMD = Unet()

        model_FMD.load_state_dict(torch.load('./models/FMD_epoch50_model'))

        loss_function = MSELoss()
        optimizer_random = Adam(model_random.parameters(), lr=0.001)
        optimizer_FMD = Adam(model_FMD.parameters(), lr=0.0001)

        model_random = model_random.to(device)
        model_FMD = model_FMD.to(device)

        #print("Training: model initialized with FMD dataset training")
        model_FMD = train(model_FMD, loader, loss_function, optimizer_FMD, n_epoch, \
                          masker, earlystop=True, patience=10, device=device, verbose=False)
        #print("Training: model randomly initialized")
        model_random = train(model_random, loader, loss_function, optimizer_random, n_epoch, \
                             masker, earlystop=True, patience=10, device=device, verbose=False)

        output_FMD = model_FMD(noisy)
        output_random = model_random(noisy)

        output_FMD = output_FMD.cpu().detach().numpy()
        output_random = output_random.cpu().detach().numpy()
        noisy = noisy.cpu().detach().numpy()
        clean = clean.cpu().detach().numpy()

        for sample in range(sample_size):
            output_random[sample,:] = match_intensity(clean[sample,:], output_random[sample,:])
            output_FMD[sample,:] = match_intensity(clean[sample,:], output_FMD[sample,:])

        quantify(FMD[key_s], metrics_key, clean[0,0, :], output_FMD[0,0,:])
        quantify(random[key_s], metrics_key, clean[0,0, :], output_random[0,0,:])
```
