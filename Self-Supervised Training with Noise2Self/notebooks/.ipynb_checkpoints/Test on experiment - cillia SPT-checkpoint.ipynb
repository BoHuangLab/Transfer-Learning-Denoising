# Noise2self training

```python
patience = 30

for frame in range(nframe):
    losses = []
    val_losses = []
    best_images = []
    best_val_loss = 1
    iters_no_improve = 0
    
    
    print("Frame:", frame)
    #noisy = data[frame,:] * sgmask
    #bg = data[frame,:] - data[frame,:] * sgmask
    
    model = Unet()
    model.load_state_dict(torch.load('./models/FMD_epoch50_model_nearestup'))

    optimizer = Adam(model.parameters(), lr=0.001)
    loss_function = MSELoss()

    model = model.to(device)

    for i in range(10000):
        model.train()
    
        net_input, mask = masker.mask(noisy, i % (masker.n_masks - 1))
        net_output = model(net_input)
    
        loss = loss_function(net_output*mask, noisy*mask) * masker.n_masks
        optimizer.zero_grad()
 
        loss.backward()
        losses.append(loss.item())
    
        optimizer.step()
    
        model.eval()
        
        net_input, mask = masker.mask(noisy, masker.n_masks - 1)
        net_output = model(net_input)
    
        val_loss = loss_function(net_output*mask, noisy*mask) * masker.n_masks
        
        val_losses.append(val_loss.item())
        
        if i % 10 == 0:
            print("(", i, ") Loss: \t", round(loss.item(), 5), "\tVal Loss: \t", round(val_loss.item(), 5))

        if val_loss.item() < best_val_loss:
            best_val_loss = val_loss.item()
            denoised = np.clip(model(noisy).detach().cpu().numpy()[0, 0], 0, 1).astype(np.float64)
            best_images.append(denoised)
            iters_no_improve = 0
        else:
            iters_no_improve += 1
            if iters_no_improve == patience:
                print("early stopping at %d iterations!" % (i))
                break
                
    output[frame,:] = best_images[-1] #+ bg.detach().cpu().numpy()
```
