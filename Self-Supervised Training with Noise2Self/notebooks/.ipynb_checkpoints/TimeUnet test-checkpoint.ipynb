# Masking

The strategy is to train a $J$-invariant version of a neural net by replacing a grid of pixels with the average of their neighbors, then only evaluating the model on the masked pixels.

```python
from mask import Masker
masker = Masker(width = 4, mode='interpolate')
```

# Model

For our model, we use a deep convolutional network.

```python
torch.manual_seed(3)
from models.timeunet import TimeUnet

nshots = 11
model = TimeUnet(time_steps=nshots)
model = model.to(device)
```

# Training

```python
from torch.nn import MSELoss, L1Loss
from torch.optim import Adam

loss_function = MSELoss()
optimizer = Adam(model.parameters(), lr=0.001)
```

```python
net_input = torch.tensor(movie[0:nshots,:,:])
net_input = net_input.to(device)
net_input = net_input.unsqueeze(0)
net_input = net_input.unsqueeze(0)
```

```python
n_middle = nshots // 2
image = movie[n_middle,:,:]
```

```python
losses = []
val_losses = []
best_images = []
best_val_loss = 1

for i in range(500):
    model.train()
    
    middle_frame = net_input[:,:,n_middle,:,:]
    masked_middle_frame, mask = masker.mask(middle_frame, i % (masker.n_masks - 1))
    net_input[:,:,n_middle,:,:] = middle_frame = masked_middle_frame
    net_output = model(net_input)
    
    loss = loss_function(net_output*mask, middle_frame*mask)
    optimizer.zero_grad()
 
    loss.backward()
    
    optimizer.step()
    
    if i % 10 == 0:
        losses.append(loss.item())
        model.eval()
        
        middle_frame = net_input[:,:,n_middle,:,:]
        masked_middle_frame, mask = masker.mask(middle_frame, i % (masker.n_masks - 1))
        net_input[:,:,n_middle,:,:] = middle_frame = masked_middle_frame
        net_output = model(net_input)
    
        val_loss = loss_function(net_output*mask, middle_frame*mask)
        
        val_losses.append(val_loss.item())
        
        print("(", i, ") Loss: \t", round(loss.item(), 5), "\tVal Loss: \t", round(val_loss.item(), 5))

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            denoised = np.clip(model(net_input).detach().cpu().numpy()[0, 0], 0, 1).astype(np.float64)
            best_psnr = compare_psnr(denoised, image)
            best_images.append(denoised)
            print("\tModel PSNR: ", np.round(best_psnr, 2))
```
